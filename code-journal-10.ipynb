```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n =1000

x = np.random.poisson(200,n)


width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x))

bins = np.arange(histmin,histmax,width)
plt.hist(x, bins = 100, color='yellow', alpha=0.75, edgecolor='grey')
plt.ylabel('values')
plt.xlabel('frequency')
plt.savefig('cj10.pdf',bbox_inches="tight",dpi=600)


```
