```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
print(x)
print(np.exp(x))
```

```python
y = np.exp(x)
```

```python
plt.plot(x,y)
plt.xlabel('Time[milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Awesomeness over time')
plt.savefig("Awesomeness.pdf",bbox_inches="tight",dpi = 600)
```
